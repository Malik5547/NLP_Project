```python
def parse_json(file_path):

  # Step 2: Open the file in read mode
  try:
    with open(file_path, "r") as json_file:
      # Step 3: Load the JSON data using json.load()
      parsed_file = json.load(json_file)
  except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
  else:
    print("JSON data parsed successfully!")
    # Step 4: Access and process the data
    # (See examples below based on data structure)
  return parsed_file
```

```python
training_set = parse_json("./train_data.json")
test_set = parse_json("./test_data.json")
```

```python
text_words_info = []
nlp = spacy.load('es_core_news_md')  # Load the Spanish language model
for i in range(len(training_set)):
  text = training_set[i]['data']['text']

  # Parse the text
  doc = nlp(text)

  # Extract sentences
  sentences = [sent.text for sent in doc.sents]

  # Extract words with start and end positions, lemma, and POS tagging for each word
  words_info = [{"sent_idx":idx, "sent":sent,"word":token.text, "start":token.idx, "end":token.idx + len(token), "lemma":token.lemma_, "pos":token.pos_,"tag":' ',"prefix":token.prefix_,"suffix":token.suffix_}
                for idx,sent in enumerate(doc.sents) for token in sent if not token.is_space and token.text not in ["*", "(",",", "?",",","!",":",";",".","&","\",","/","-"]]
  text_words_info.append(words_info)
```

```python
text_words_info_test = []
nlp = spacy.load('es_core_news_md')  # Load the Spanish language model
for i in range(len(test_set)):
  text = test_set[i]['data']['text']

  # Parse the text
  doc = nlp(text)

  # Extract sentences
  sentences = [sent.text for sent in doc.sents]

  # Extract words with start and end positions, lemma, and POS tagging for each word
  words_info = [{"sent_idx":idx, "sent":sent,"word":token.text, "start":token.idx, "end":token.idx + len(token), "lemma":token.lemma_, "pos":token.pos_,"tag":' ',"prefix":token.prefix_,"suffix":token.suffix_}
                for idx,sent in enumerate(doc.sents) for token in sent if not token.is_space and token.text not in ["*", "(",",", "?",",","!",":",";",".","&","\",","/","-"]]
  text_words_info_test.append(words_info)
```

Create the vocabulary of words for the training set

```python
training_vocab = set()
training_words_list = []

for doc in text_words_info:
  for word_info in doc:
    word = word_info["word"]
    if word.isalpha():
      training_words_list.append(word)

training_vocab = set(training_words_list)

print(len(training_vocab))   
print(training_vocab) 
print(len(training_words_list))
```

Create the vocabulary of words for the test set

```python
test_vocab = set()
test_words_list = []

for doc in text_words_info_test:
  for word_info in doc:
    word = word_info["word"]
    if word.isalpha():
      test_words_list.append(word)

test_vocab = set(test_words_list)

print(len(test_vocab))   
print(test_vocab) 
print(len(test_words_list))
```

Checking how many words from test set are present in train set.

```python
present = 0
for word in test_vocab:
    if word in training_vocab:
        present += 1

print(present)
```

```python
present = 0
for word in test_words_list:
    if word in training_vocab:
        present += 1

print(present)
```

```python
training_vocab_lemma = set()
training_words_list_lemma = []

for doc in text_words_info:
  for word_info in doc:
    word = word_info["word"]
    if word.isalpha():
      training_words_list_lemma.append(word_info['lemma'])

training_vocab_lemma = set(training_words_list_lemma)

print(len(training_vocab_lemma))   
print(training_vocab_lemma) 
print(len(training_words_list_lemma))
```

```python
test_vocab_lemma = set()
test_words_list_lemma = []

for doc in text_words_info_test:
  for word_info in doc:
    word = word_info["word"]
    if word.isalpha():
      test_words_list_lemma.append(word_info['lemma'])

test_vocab_lemma = set(test_words_list_lemma)

print(len(test_vocab_lemma))   
print(test_vocab_lemma) 
print(len(test_words_list_lemma))
```

```python
present = 0
new_words = []
for word in test_vocab_lemma:
    if word in training_vocab_lemma:
        present += 1
    else :
        new_words.append(word)

print(present)
print(new_words)
```
